## 1. Feature Selection Methodologies

### IV-Based Selection (Production Model)
- **Method**: OptBinning Weight of Evidence (WoE) and Information Value (IV)
- **Approach**: Univariate statistical analysis on each feature
- **Criteria**: Features with IV >= 0.02 and < 1.0 (no data leakage)
- **Advantage**: Fast, interpretable, industry standard for credit risk
- **Limitation**: Single-variable analysis, doesn't consider multivariate relationships

### RFECV Selection (Validation Approach)
- **Method**: Recursive Feature Elimination with Cross-Validation (RFECV)
- **Estimator**: Explainable Boosting Trees (EBT)
- **CV Strategy**: 5-Fold Stratified Cross-Validation
- **Scoring Metric**: ROC-AUC
- **Advantage**: Multivariate analysis, accounts for feature interactions
- **Limitation**: Computationally intensive, black-box optimization

## 8. Feature Engineering Pipeline

Both approaches use the same time-series features from feature_engineering_timeseries.py:

### Rolling Window Features (7d, 30d, 90d, 180d)
- `avg_proc_vol_{window}`: Average processing volume
- `min_proc_vol_{window}`: Minimum processing volume
- `max_proc_vol_{window}`: Maximum processing volume
- `std_proc_vol_{window}`: Std deviation processing volume
- `cv_proc_vol_{window}`: Coefficient of variation (volatility)
- `avg_tx_hours_{window}`: Average transaction hours
- Similar for Tx_hours metrics

### Momentum Features
- `pct_change_proc_vol_vs_{window}_ago`: % change vs lag dates
- `pct_change_tx_hours_vs_{window}_ago`: % change vs lag dates

### Temporal Features
- `snapshot_day_of_week`: Day of week (1-7)
- `snapshot_month`: Month (1-12)
- `snapshot_quarter`: Quarter (1-4)

### Restaurant-Level Features
- Account features from Lending_default_train_account.csv
- Categorical: Ownership_type, Restaurant_category, Market_segment
- Numeric: Various account metrics
